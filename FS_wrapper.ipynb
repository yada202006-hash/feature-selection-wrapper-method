# Wrapper method
It is a type of feature selection methods that involve a predictive model to score the combination of features

# steps
1. subset genration
2. subset evaluation
3. stopping criteria

# techniques used
1. Exhaustive feature selection
2. forward selection
3. backward elimination
4. recursive feature elimination

# Exhaustive feature selection

 A wrapper‑based technique that tries every possible subset of features and selects the one yielding the best model performance

 ex- if we have 3 columns f1,f2,f3
 we try all combination f1,f2,f3,f1f2,f1f3,f2f3

**This method have many flaws**
like **computational complexity**(we have to train 2^𝑛-1 models)
*Risk of overfitting*

# Backward elimination
Backward Elimination is a stepwise feature selection method that starts with all available features and iteratively removes the least significant ones until only the most relevant remain. It’s commonly used in regression modeling but can be applied more broadly.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
import pandas as pd
import numpy as np
```

```python
# load dataset
data = pd.read_csv(r"C:\Users\Sonu\OneDrive\Desktop\machine learning\feature selction\BostonHousing.csv")

data.fillna(data.mean(), inplace = True)

x=data.drop("medv",axis=1)
y=data["medv"]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)

print(x_test.shape,x_train.shape)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
import numpy as np

model = LinearRegression()
train_scores = cross_val_score(model, x_train, y_train, cv=5, scoring='r2')
print("Training R²:", np.mean(train_scores))

print("testing",np.mean(cross_val_score(model,x_test,y_test,cv=5,scoring="r2")))
```

```python
lr=LinearRegression()
# perform backward elimination
sfs=SFS(lr,k_features="best",forward=False,floating=False,scoring='r2',cv=5)
sfs.fit(x_train,y_train)
```

```python
sfs.k_feature_idx_
```

```python
x_train_sel=sfs.transform(x_train)
x_test_sel=sfs.transform(x_test)
```

```python
model=LinearRegression()
print("training",np.mean(cross_val_score(model,x_train_sel,y_train,cv=5,scoring='r2')))
print("testing",np.mean(cross_val_score(model,x_test_sel,x_test,cv=5,scoring='r2')))
```

# seuential forward selection
Forward Sequential Selection is a wrapper feature selection method that builds the model step‑by‑step by adding features one at a time. It’s the opposite of backward elimination.

```python
lr=LinearRegression()
# perform backward elimination
sfs=SFS(lr,k_features=9,forward=True,floating=False,scoring='r2',cv=5)
sfs.fit(x_train,y_train)
```

```python
sfs.k_feature_idx_
```

```python
x_train_sel=sfs.transform(x_train)
x_test_sel=sfs.transform(x_test)
```

```python
model=LinearRegression()
print("training",np.mean(cross_val_score(model,x_train_sel,y_train,cv=5,scoring='r2')))
print("testing",np.mean(cross_val_score(model,x_test_sel,x_test,cv=5,scoring='r2')))
```
